```python
from transformers import pipeline

chatbot = pipeline("text-generation", model="gpt2")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Bye!")
        break

    response = chatbot(
        user_input,
        max_new_tokens=60,
        do_sample=True,
        temperature=0.7
    )

    reply = response[0]['generated_text'][len(user_input):]

    print("Chatbot:", reply)
```
